trin1 <- sense %>% enframe(name=NULL) %>% 
  mutate(chapter = case_when(
    str_detect(value, "CHAPTER \\d") ~ value,
    .default = NA_character_
  )) %>% 
  fill(chapter, .direction = "down") %>% 
  filter(value != chapter) %>% 
  group_by(chapter) %>% 
  summarise(text= paste0(value, collapse= " ")) %>% 
  ungroup() %>% 
  mutate(chapter = as.numeric(str_remove(chapter, "CHAPTER")))   %>% 
  mutate(sentences = map(text, spacy_tokenize, 'sentence'))

  unnest_longer(sentences) %>%
